2.1.2. Построить фукнцию распределения степеней (в обычных и логарифимических (по 1й и по 2м осям) системах координат) узлов для визуализированных графиков.

```python
def cal_deg_cnt(G):
    degree_sequence = sorted([d for n, d in G.degree()])
    degreeCount = collections.Counter(degree_sequence)
    return zip(*degreeCount.items())

def plot_hist(G):
    deg, cnt = cal_deg_cnt(G)

    fig, ax = plt.subplots(figsize=(20,10))
    ax.bar(deg, cnt, width=0.80, color='b')
    plt.title("Degree Histogram")
    plt.ylabel("Count")
    plt.xlabel("Degree")
    ax.set_xticks([d + 0.4 for d in deg])
    ax.set_xticklabels(deg)
    degree_sequence = sorted([d for n, d in G.degree()])

def plot_scatter(G):
    deg, cnt = cal_deg_cnt(G)

    # обычная система координат
    fig, ax = plt.subplots(figsize=(20,10))
    ax.scatter(deg, cnt, color='b', marker="o")
    ax.set_xticks([d for d in deg])
    ax.set_xticklabels(deg)
    plt.title("Распределение степеней")
    plt.xlabel("Степень")
    plt.show()

def plot_log(G):
    deg, cnt = cal_deg_cnt(G)

    # логарифмическая система координат
    fig, ax = plt.subplots(figsize=(20,10))
    ax.scatter(deg, cnt, color='b', marker="o")
    plt.xscale('log')
    plt.yscale('log')
    plt.title("Распределение степеней")
    plt.xlabel("Степень")
    plt.show()
    
G = nx.read_edgelist(filePath, comments='%', nodetype=int)
plot_hist(G) # Гистаграмма распределения
plot_scatter(G) # точечный график
plot_log(G) # логорифмизированный масштаб 
```

2.1.3. Построить кумулятивный график распределения.

```python
# кумулятивный график распределения
def cal_cum_prob(cnt):
    n = sum(cnt)
    probs = [i/n for i in cnt]
    cum_prob = []
    prob_sum = 0
    for prob in probs:
        prob_sum += prob
        cum_prob.append(prob_sum)
    return cum_prob

def plot_cum(G):
    deg, cnt = cal_deg_cnt(G);
    cum_prob = cal_cum_prob(cnt)
    fig, ax = plt.subplots(figsize=(20,10))
    plt.step(deg, cum_prob, color='b', marker="o")
    ax.set_xticks([d for d in deg])
    ax.set_xticklabels(deg)
    plt.title("Distribution")
    plt.xlabel("Degree")
    plt.show()
    
plot_cum(G)
```

2.1.4. Построить распределение с log-binning.

```python
# автор: Андрей Бахматов
def log_binning(G,bin_count=35,original=False):
    ungrouped_deg = dict(G.degree())
    degrees = dict(collections.Counter(ungrouped_deg.values()))
    
    keys = list((degrees.keys()))
    values = list(degrees.values())
    
    max_x = np.log10(max(keys))
    max_y = np.log10(max(values))
    max_base = max([max_x,max_y])
    
    min_x = np.log10(min(keys))

    bins = np.logspace(min_x,max_base,num=bin_count)
    # среднее значение в каждом интервале, для обоих осей
    bin_means_x = np.histogram(keys, bins, weights=keys)[0] / np.histogram(keys, bins)[0]
    bin_means_y = np.histogram(keys, bins, weights=values)[0] / np.histogram(keys, bins)[0]
    
    fig, ax = plt.subplots(figsize=(20,10))
    if original:
        deg, cnt = zip(*degrees.items())
        ax.scatter(deg, cnt, color='b', marker="o", label="Изначальное распределение")
        ax.set_xticks([d for d in deg])
        ax.set_xticklabels(deg)
        
    ax.scatter(bin_means_x,bin_means_y,c='r',marker='s',s=50,alpha=0.7,label='log-binning');
    plt.xscale('log');
    plt.yscale('log');
    plt.xlabel('Степень');
    plt.ylabel('Частота');
    plt.legend();
    plt.show()

log_binning(G,bin_count=7,original=True)
```
